# Ejemplo imagenes (borrar)

```python
# BORRAR
# Install the pooch package
%pip install pooch

from skimage import data, img_as_float, io, color
from skimage.util import invert
import matplotlib.pyplot as plt

image1 = img_as_float(data.cell())
image2 = img_as_float(data.horse())
image3 = img_as_float(data.text())
image3 = img_as_float(data.rocket())
imagen_gris = color.rgb2gray(image3)

# Paso 2: Convertir cada imagen a su negativo
negative1 = invert(image1)
negative2 = invert(image2)
negative3 = invert(image3)
negative4 = invert(imagen_gris)

# Paso 3: Mostrar las imágenes originales y procesadas
fig, axes = plt.subplots(4, 2, figsize=(10, 15))

# Mostrar imagen 1
axes[0, 0].imshow(image1, cmap="gray")
axes[0, 0].set_title("Imagen Original 1")
axes[0, 0].axis("off")

axes[0, 1].imshow(negative1, cmap="gray")
axes[0, 1].set_title("Imagen Negativa 1")
axes[0, 1].axis("off")

# Mostrar imagen 2
axes[1, 0].imshow(image2, cmap="gray")
axes[1, 0].set_title("Imagen Original 2")
axes[1, 0].axis("off")

axes[1, 1].imshow(negative2, cmap="gray")
axes[1, 1].set_title("Imagen Negativa 2")
axes[1, 1].axis("off")

# Mostrar imagen 3
axes[2, 0].imshow(image3, cmap="gray")
axes[2, 0].set_title("Imagen Original 3")
axes[2, 0].axis("off")

axes[2, 1].imshow(negative3, cmap="gray")
axes[2, 1].set_title("Imagen Negativa 3")
axes[2, 1].axis("off")

# Mostrar imagen 4
axes[3, 0].imshow(imagen_gris, cmap="gray")
axes[3, 0].set_title("Imagen Original 3")
axes[3, 0].axis("off")

axes[3, 1].imshow(negative4, cmap="gray")
axes[3, 1].set_title("Imagen Negativa 3")
axes[3, 1].axis("off")



plt.tight_layout()
plt.show()
```

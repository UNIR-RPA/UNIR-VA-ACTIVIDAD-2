```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io, morphology
from skimage.morphology import disk
from skimage.filters import gaussian  # Filtro Gaussiano

# Cargar la imagen desde el archivo subido
image = io.imread('radiografia.jpg')

# Aplicar un filtro Gaussiano para suavizar la imagen
image_gaussian = gaussian(image, sigma=3, preserve_range=True)  # Suavizado con filtro Gaussiano

# Binarización sin filtro Gaussiano
threshold = 160
binary_without_gaussian = image > threshold  # Binarización directa sobre la imagen original

# Binarización con filtro Gaussiano
binary_with_gaussian = image_gaussian > threshold  # Aplicamos la binarización

# Aplicar dilatación seguida de erosión
dilation_result = morphology.dilation(binary_with_gaussian, footprint=disk(10))  # Dilatación
final_result = morphology.erosion(dilation_result, footprint=disk(18))  # Erosión

# Mostrar las imágenes lado a lado
fig, ax = plt.subplots(1, 6, figsize=(24, 10))

# Imagen original
ax[0].imshow(image, cmap='gray')
ax[0].set_title("Imagen Original")
ax[0].axis('off')

# Imagen después del filtro Gaussiano
ax[1].imshow(image_gaussian, cmap='gray')
ax[1].set_title("Filtro Gaussiano")
ax[1].axis('off')

# Imagen binarizada sin filtro Gaussiano
ax[2].imshow(binary_without_gaussian, cmap='gray')
ax[2].set_title("Binarización sin Gaussiano")
ax[2].axis('off')

# Imagen binarizada con filtro Gaussiano
ax[3].imshow(binary_with_gaussian, cmap='gray')
ax[3].set_title("Binarización con Gaussiano")
ax[3].axis('off')

# Imagen tras la dilatación
ax[4].imshow(dilation_result, cmap='gray')
ax[4].set_title("Tras Dilatación")
ax[4].axis('off')

# Imagen final tras dilatación y erosión
ax[5].imshow(final_result, cmap='gray')
ax[5].set_title("Tras Erosión")
ax[5].axis('off')

plt.tight_layout()
plt.show()
```

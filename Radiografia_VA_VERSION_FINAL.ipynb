```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io, morphology
from skimage.morphology import disk
from skimage.filters import gaussian, threshold_otsu  # Filtro Gaussiano y Otsu

# Cargar la imagen en escala de grises
image = io.imread('radiografia.jpg', as_gray=True)

# Aplicar un filtro Gaussiano para suavizar la imagen
image_gaussian = gaussian(image, sigma=3, preserve_range=True)  # Suavizado con filtro Gaussiano

# Umbrales automáticos (Otsu) para cada caso
threshold_orig = threshold_otsu(image)
threshold_gauss = threshold_otsu(image_gaussian)

# Binarización usando umbral Otsu
binary_without_gaussian = image > threshold_orig
binary_with_gaussian = image_gaussian > threshold_gauss

# Aplicar dilatación seguida de erosión
dilation_result = morphology.dilation(binary_with_gaussian, footprint=disk(10))  # Dilatación
final_result = morphology.erosion(dilation_result, footprint=disk(18))  # Erosión

# Mostrar las imágenes lado a lado
fig, ax = plt.subplots(1, 6, figsize=(24, 10))

# Imagen original
ax[0].imshow(image, cmap='gray')
ax[0].set_title("Imagen Original")
ax[0].axis('off')

# Imagen después del filtro Gaussiano
ax[1].imshow(image_gaussian, cmap='gray')
ax[1].set_title("Filtro Gaussiano")
ax[1].axis('off')

# Imagen binarizada sin filtro Gaussiano
ax[2].imshow(binary_without_gaussian, cmap='gray')
ax[2].set_title("Binarización sin Gaussiano (Otsu)")
ax[2].axis('off')

# Imagen binarizada con filtro Gaussiano
ax[3].imshow(binary_with_gaussian, cmap='gray')
ax[3].set_title("Binarización con Gaussiano (Otsu)")
ax[3].axis('off')

# Imagen tras la dilatación
ax[4].imshow(dilation_result, cmap='gray')
ax[4].set_title("Tras Dilatación")
ax[4].axis('off')

# Imagen final tras dilatación y erosión
ax[5].imshow(final_result, cmap='gray')
ax[5].set_title("Tras Erosión")
ax[5].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io, morphology, exposure
from skimage.morphology import disk
from skimage.util import img_as_float

# Cargar imagen a color (RGB) y convertir a float [0,1]
image_rgb = io.imread('f1.jpg')
image = img_as_float(image_rgb)

# 1) Erosión por canal para eliminar detalles brillantes
se_erode = disk(3)  # Ajustable según tamaño de detalles
eroded = np.empty_like(image)
for c in range(3):
    eroded[..., c] = morphology.erosion(image[..., c], footprint=se_erode)

# 2) Gradiente interno (original - erosionada) por canal
internal_grad = image - eroded

# 3) Dilatación por canal con disco de tamaño 1
se_dil = disk(1)
dilated_grad = np.empty_like(internal_grad)
for c in range(3):
    dilated_grad[..., c] = morphology.dilation(internal_grad[..., c], footprint=se_dil)

# Reescalar intensidades para mejor visualización
internal_grad_v = exposure.rescale_intensity(internal_grad, out_range=(0, 1))
dilated_grad_v = exposure.rescale_intensity(dilated_grad, out_range=(0, 1))

# Mostrar resultados
fig, ax = plt.subplots(1, 4, figsize=(18, 6))
ax[0].imshow(image)
ax[0].set_title('Original (RGB)')
ax[0].axis('off')

ax[1].imshow(eroded)
ax[1].set_title('Erosionada (RGB)')
ax[1].axis('off')

ax[2].imshow(internal_grad_v)
ax[2].set_title('Gradiente interno (RGB)')
ax[2].axis('off')

ax[3].imshow(dilated_grad_v)
ax[3].set_title('Gradiente dilatado (RGB, disk=1)')
ax[3].axis('off')

plt.tight_layout()
plt.show()
```
